```python
import tkinter as tk
from tkinter import messagebox

# Sample account data (You should use a database for real-world applications)
accounts = {
    '123456': {'pin': '1234', 'balance': 1000},
    '789012': {'pin': '5678', 'balance': 1500}
}

current_account = None

def login():
    global current_account
    account_number = account_entry.get()
    pin = pin_entry.get()

    if account_number in accounts and pin == accounts[account_number]['pin']:
        current_account = account_number
        balance_label.config(text=f'Balance: ${accounts[current_account]["balance"]}')
        main_menu()
    else:
        messagebox.showerror('Error', 'Invalid Account Number or PIN')

def main_menu():
    root.withdraw()  # Hide the login window
    main_menu_window = tk.Toplevel(root)
    main_menu_window.title('ATM Simulator')

    def check_balance():
        balance = accounts[current_account]['balance']
        messagebox.showinfo('Balance', f'Your balance is: ${balance}')

    def deposit():
        deposit_window = tk.Toplevel(main_menu_window)
        deposit_window.title('Deposit')
        amount_label = tk.Label(deposit_window, text='Enter deposit amount:')
        amount_label.pack()
        deposit_entry = tk.Entry(deposit_window)
        deposit_entry.pack()

        def deposit_money():
            try:
                amount = float(deposit_entry.get())
                if amount > 0:
                    accounts[current_account]['balance'] += amount
                    balance_label.config(text=f'Balance: ${accounts[current_account]["balance"]}')
                    deposit_window.destroy()
                else:
                    messagebox.showerror('Error', 'Invalid amount')
            except ValueError:
                messagebox.showerror('Error', 'Invalid amount')

        deposit_button = tk.Button(deposit_window, text='Deposit', command=deposit_money)
        deposit_button.pack()

    def withdraw():
        withdraw_window = tk.Toplevel(main_menu_window)
        withdraw_window.title('Withdraw')
        amount_label = tk.Label(withdraw_window, text='Enter withdrawal amount:')
        amount_label.pack()
        withdraw_entry = tk.Entry(withdraw_window)
        withdraw_entry.pack()

        def withdraw_money():
            try:
                amount = float(withdraw_entry.get())
                if amount > 0 and amount <= accounts[current_account]['balance']:
                    accounts[current_account]['balance'] -= amount
                    balance_label.config(text=f'Balance: ${accounts[current_account]["balance"]}')
                    withdraw_window.destroy()
                else:
                    messagebox.showerror('Error', 'Invalid amount or insufficient balance')
            except ValueError:
                messagebox.showerror('Error', 'Invalid amount')

        withdraw_button = tk.Button(withdraw_window, text='Withdraw', command=withdraw_money)
        withdraw_button.pack()

    check_balance_button = tk.Button(main_menu_window, text='Check Balance', command=check_balance)
    deposit_button = tk.Button(main_menu_window, text='Deposit', command=deposit)
    withdraw_button = tk.Button(main_menu_window, text='Withdraw', command=withdraw)

    check_balance_button.pack()
    deposit_button.pack()
    withdraw_button.pack()

root = tk.Tk()
root.title('ATM Simulator')

account_label = tk.Label(root, text='Enter Account Number:')
account_label.pack()
account_entry = tk.Entry(root)
account_entry.pack()

pin_label = tk.Label(root, text='Enter PIN:')
pin_label.pack()
pin_entry = tk.Entry(root, show='*')
pin_entry.pack()

login_button = tk.Button(root, text='Login', command=login)
login_button.pack()

balance_label = tk.Label(root, text='Balance: $0')
balance_label.pack()

root.mainloop()
```
